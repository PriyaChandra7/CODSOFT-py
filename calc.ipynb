```python
#calculator
from tkinter import *

def buttonClick(number):
    global val
    val=val+str(number)
    data.set(val)

def buttonClear():
    global val
    val=" "
    data.set("")
    
def buttonEqual():
    global val
    result=eval(val) #eval=evaluates a string as a Python expression
    data.set(result)
    
window=Tk()
window.title("My Calculator")
window.geometry("360x380")  # width x height +x +y
val=""
data=StringVar() #stringvar= dynamically update the content displayed in Entry.


display=Entry(window,textvariable=data,bd=29,justify="right",bg="light green",font=("ariel",20))

#to get the value outside the display use another variable - textvariable named as data
#textvariable is an attribute(variables/func)
#When you assign a StringVar to the textvariable attribute of a widget like Entry, changes to the tkinter variable (StringVar)
#automatically update the displayed content of the widget.
#bd=border 
#justify option is specifically used to align the text inside an Entry widget.
#display.config(bd=29,bg="grey")  use if we need to change this dynamically
#display.pack() in pack we can't use multiple columns 

display.grid(row=0,columnspan=4) #Places the Entry widget in the first row, spanning all 4 columns.

#we can only pass argumentt using lambda function Without lambda, you would not be able to pass specific arguments to the function
button7=Button(window,text="7",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick(7)) 
button7.grid(row=1,column=0) #use padx and pady inside grid
button8=Button(window,text="8",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick(8))   
#when click button digit display on screen so we need to call function
button8.grid(row=1,column=1)
button9=Button(window,text="9",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick(9))
button9.grid(row=1,column=2)
buttonAdd=Button(window,text="+",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick('+'))
buttonAdd.grid(row=1,column=3)

button4=Button(window,text="4",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick(4))
button4.grid(row=2,column=0)
button5=Button(window,text="5",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick(5))
button5.grid(row=2,column=1)
button6=Button(window,text="6",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick(6))
button6.grid(row=2,column=2)
buttonSub=Button(window,text="-",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick('-'))
buttonSub.grid(row=2,column=3)

button1=Button(window,text="1",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick(1))
button1.grid(row=3,column=0)
button2=Button(window,text="2",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick(2))
button2.grid(row=3,column=1)
button3=Button(window,text="3",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick(3))
button3.grid(row=3,column=2)
buttonMul=Button(window,text="*",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick('*'))
buttonMul.grid(row=3,column=3)

buttonCl=Button(window,text="C",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=buttonClear)
buttonCl.grid(row=4,column=0)
button0=Button(window,text="0",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick(0))
button0.grid(row=4,column=1)
buttonDiv=Button(window,text="/",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=lambda:buttonClick('/'))
buttonDiv.grid(row=4,column=2)
buttonEqual=Button(window,text="=",font=("ariel",12,"bold"),bd=12,height=2,width=6,command=buttonEqual) #
buttonEqual.grid(row=4,column=3)
window.mainloop() 
```
