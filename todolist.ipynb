```python
#to do list gui project

from tkinter import *
from tkinter import messagebox #to display a message

def newTask():
    task= entry.get()
    if task != "":
        lb.insert(END,task)
        entry.delete(0, "end") #the entry box will go empty when we click the button addtask 0 to end is range 
    else:
        messagebox.showwarning("warning","please enter task")
        

def deleteTask():
    lb.delete(ANCHOR) #anchor refer to the currently selected items

window=Tk()
window.geometry("380x390") #widthxheight 
window.title("YOUR TO DO LIST :)")

#create widgets(frame, listbox, scrollbar,entry,button)

frame= Frame(window)
frame.pack(pady=10)

lb=Listbox(frame,bg="light blue",width=25,height=8,font=('times',16))
lb.pack(side=LEFT,fill=BOTH)

task_list=[
    'Bring grocerries',
    'Drink water',
    'go to gym',
    '5:00 pm meeting with a client',
    'go for a party',
    'bring some flower at evening',
    'wash dishes',
    ]
for item in task_list:
    lb.insert(END, item)

sb=Scrollbar(frame)
sb.pack(side=RIGHT,fill=BOTH)

#this is impt
lb.config(yscrollcommand=sb.set)
sb.config(command=lb.yview)

entry=Entry(window,font=("times",18),bg="grey",bd=2)
entry.pack(pady=20)

button_frame=Frame(window)
button_frame.pack(pady=20)

addTask_btn= Button(button_frame,text="Add Task",font=("times",15),bg="light green",command=newTask)
addTask_btn.pack(side=LEFT,fill=BOTH)

addDel_btn=Button(button_frame,text="Delete Task",font=("times",15),bg="yellow",command=deleteTask)
addDel_btn.pack(side=RIGHT,fill=BOTH)


window.mainloop() #we can see the window because of this infinite loop running 
```
